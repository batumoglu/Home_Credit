```python
import matplotlib.pyplot as plt
import seaborn as sns
for column in credit_card.columns:
    credit_card[column] = credit_card[column].fillna(0)
    if(credit_card[column].dtypes!='object'):
        try:
            plt.figure(figsize=(12,8))
            type(credit_card[column])
            sns.distplot(credit_card[column], bins=50, kde=False)
            plt.show()
        except:
            print('There is NaN data in column: %s' %column)
            plt.show()
```

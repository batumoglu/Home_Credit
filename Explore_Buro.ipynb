```python
import matplotlib.pyplot as plt
import seaborn as sns
for column in buro.columns:
    buro[column] = buro[column].fillna(0)
    if(buro[column].dtypes!='object'):
        try:
            plt.figure(figsize=(12,8))
            type(buro[column])
            sns.distplot(buro[column], bins=50, kde=False)
            plt.show()
        except:
            print('There is NaN data in column: %s' %column)
            plt.show()
```
